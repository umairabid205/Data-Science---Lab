```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('/Users/mh/Desktop/DS_Task/loangrant.csv')
df.head()
```

```python


# Assuming your DataFrame is named 'df', and you want to calculate confidence intervals for all numeric columns
numeric_columns = df.select_dtypes(include='number').columns

confidence_intervals = {}

for column in numeric_columns:
    data = df[column].dropna()  # Remove missing values
    confidence_interval = stats.t.interval(
        0.95, len(data) - 1, loc=data.mean(), scale=stats.sem(data))
    confidence_intervals[column] = confidence_interval

# Display the confidence intervals
for column, interval in confidence_intervals.items():
    print(f"Confidence Interval (95%) for {column}: {interval}")
```
